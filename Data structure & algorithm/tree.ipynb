- 후위순회(LRV) : 폴더 용량계산 -> 모든 파일을 다 돌고 root폴더에 도달

### 개수와 높이
- 노드의 개수 : 왼쪽 트리의 노드수 + 오른쪽 트리의 노드수 + 1(자기자신)
- 단말 노드의 수 : 양쪽 트리가 모두 None인 노드일 때 카운트
- 트리의 높이 : max(왼쪽 트리의 높이,오른쪽 트리의 높이) + 1

```python
def count_node(n):
    if n is None:
        return 0 
    else:
        return count_node(n.left) + count_node(n.right) + 1

def count_leaf(n):
    if n is None:
        return 0 
    elif n.left == None and n.right == None:
        return 1
    else:
        return count_leaf(n.left) + count_leaf(n.right)
def calc_height(n):
    if n is None:
        return 0
    elif calc_height(n.left) > calc_height(n.right):
        return calc_height(n.left) + 1
    else:
        return calc_height(n.right) + 1 
```

```python
class BinaryTree:
    class Node:
        def __init__(self,item,left=None,right=None):
            self.item = item
            self.left = left
            self.right = right
    def __init__(self):
        self.root = None
    
    def preorder(self,n):
        if n is None:
            return 0 
        else:
            print(n.item,end=' ')
            self.preorder(n.left)
            self.preorder(n.right)
    def inorder(self,n):
        if n is None:
            return 0 
        else:
            self.inorder(n.left)
            print(n.item,end= ' ')
            self.inorder(n.right)
    
    def postorder(self,n):
        if n is None:
            return 0
        else:
            self.postorder(n.left)
            self.postorder(n.right)
            print(n.item,end=' ')
    
    def levelorder(self,n):
        q = []
        q.append(n)
        while len(q) != 0:
            t = q.pop(0)
            print(t.item,end=' ')
           
            if t.left != None:
                q.append(t.left)
            if t.right != None:
                q.append(t.right)
    
    def height(self,root):
        if root == None:
            return 0
        else:
            return max(self.height(root.left),self.height(root.right)) + 1
```

### 힙(HEAP)
- 완전이진트리 기반
- 힙은 반정렬 데이터(형제끼리는 비교 안됨)
- 스택과 큐 기준의 우선순위가 아닌 임의의 우선순위를 지정한 트리 -> 최대/최소 힙
- 배열을 이용하여 구현
- 한 번 수행 시 O(logN)의 시간복잡도가 걸리고 N개의 데이터에 대해선 NO(logN)이다
    - N개의 데이터에 대해선 1/2N * O(logN) 정도면 모두 정렬이 된다 
    - O(logN)이 1/2N보다 작다는 것을 가정할 때 O(N)의 시간복잡도를 가진다
#### UPheap O(logn)
- 일단 삽입 tree[-1]
- 부모(i//2)와 크기를 비교해서 자리를 바꾼다 
- 최대 루트까지 반복한다
- 장점
    -좌/우 서브트리 중 하나를 무작정 선택하면 반대쪽 트리와 크기비교가 필요없다
#### Downheap O(logN)
- 항상 루트가 삭제된다 
- 말단 노드를 루트로 올린다
- 두 자식 중에 제일 크면서 나보다 큰 노드를 올린다
- 말단이 될때까지 반복한다
- 형제끼리의 크기 비교가 특징
- 말단 노드였지만 종료시점에는 다른 노드로 바뀔 수 있다

```python
class MaxHeap:
    def __init__(self):
        self.heap = []
        self.heap.append(0)

    def size(self):
        return len(self.heap) - 1 #빈 노드 제거
    def isempty(self): return self.size() == 0
    def Parent(self,i): return self.heap[i//2]
    def Left(self,i): return self.heap[i*2]
    def Right(self,i): return self.heap[i*2+1]
    def display(self):
        print('heap:',self.heap[1:])

    def insert(self,n):
        self.heap.append(n) 
        i = self.size() #방금 들어온 노드의 위치
        while (i != 1 and n > self.Parent(i)):
            self.heap[i] = self.Parent(i)
            i = i//2
        self.heap[i] = n
    def delete(self):
        parent = 1
        child = 2 # 다음 레벨의 첫 노드
        if not self.isempty():
            hroot = self.heap[1]
            last = self.heap[-1]
            while (child <= self.size()):
                if child < self.size() and self.Left(parent) < self.Right(parent):
                    child += 1
                if last > self.heap[child]:
                    break
                self.heap[parent] = self.heap[child]
                parent = child
                child *= 2

            self.heap[parent] = last
            self.heap.pop(-1)
            return hroot
```

```python
heap = MaxHeap()

data = [2,9,9,9,3,9,7,4,3,5,6,7]
for i in data:
    heap.insert(i)
heap.display()
heap.delete()
heap.display()
```

상향식 힙
- 상향식으로 힙속성을 만족하도록 부모 자식 교환
- 임의의 순서를 가진 리스트에 대해 downheap 수행
    - 가장 하위레벨에서 downheap을 하지 않는이유
        - 자식노드가 없기 때문에 그들은 크기가 1인 최소힙이다
        - 즉, 바꿀 자식이 없어서 Downheap 대상이 아니다
- Upheap의 경우 N개의 자료를 하나씩 넣어야 해서 N번
- Downheap의 경우 일단 다 넣고 내부노드만 하니 N/2번

```python
class BHeap:
    def __init__(self,a):
        self.a = a
        self.N = len(a) - 1
    
    def create_heap(self):
        for i in range(self.N//2,0,-1):
            self.downheap(i)
    
    def insert(self,key_value):
        self.N += 1
        self.a.append(key_value)
        self.upheap(self.N)

    def delete_min(self):
        if self.N == 0:
            return None
        minimum = self.a[1]
        self.a[1] , self.a[-1] = self.a[-1], self.a[1]
        del self.a[-1]
        self.N -= 1
        self.downheap(1)
        return minimum
    
    def downheap(self,i):
        while 2*i <= self.N:
            k = 2*i
            if k<self.N and self.a[k][0] > self.a[k+1][0]:
                k += 1 
            if self.a[i][0] < self.a[k][0]:
                break
            self.a[i], self.a[k] = self.a[k], self.a[i]
            i = k
    
    def upheap(self,j):
        while j > 1 and self.a[j][0] < self.a[j//2][0]: #부모가 자식보다 더 크면 실행
            self.a[j], self.a[j//2] = self.a[j//2], self.a[j] #자리바꿈 
            j = j//2
    
    def print_heap(self):
        print(self.a)

    
```

# 탐색트리

- 순차탐색 : O(N)
- 이진탐색 : O(logN)
    - 데이터가 왼쪽과 오른쪽 중 어디에 있는가?
    - 재귀와 반복으로 모두 구현이 가능
    - Alg : T(N) = T(N/2) + 1  -> T(N/2^k) + K -> 2의 지수승만큼 N이 감소됨
- 보간탐색 : 탐색 값의 위치를 예측하여 탐색
    - 인덱스가 커질수록 탐색값도 커질 것이다

- 이진탐색트리
    - 중위순회시 정렬되어 출력됨
        - 당연하게도 루트는 리스트의 가운데에 위치해있기 때문
    - 완전이진트리일 필요는 없다(힙트리는 완전이진트리여야함)
    - 삽입, 삭제, 탐색 O(logN)
        - 트리를 내려갈때, 재연결할 때 O(1)
        - 연산시간은 레벨 수에 비례 O(h)
        - h는 완전이진트리와 편향트리 사이에 있다 logN <= h <= N ≒ log(N+1)
    - 루트를 기준으로 왼쪽은 항상 작고 오른쪽은 항상 크다
        - 모든 원소의 키는 유일한 값을 가져야한다
    ※ 루트의 근사값은 왼쪽 서브트리 중 가장 오른쪽 노드, 오른쪽 서브트리 중 가장 왼쪽 노드


```python
class BST(BinaryTree):
    class Node:
        def __init__(self,key,value,left=None,right=None):
            self.key = key
            self.item = value
            self.left = left
            self.right = right

    def __init__(self):
        self.root = None

    def get(self,key):
        return self.get_item(self.root,key)
    
    def get_item(self,n,key):
        if n == None:
            return None
        if n.key > key:
            return self.get_item(n.left,key) #키 값이 부모보다 작으면 왼쪽으로
        elif n.key < key:
            return self.get_item(n.right,key) #키 값이 부모보다 크면 오른쪽으로
        else:
            return n.value #같으면 value 반환
        
    def put(self,key,value):
        self.root = self.put_item(self.root,key,value)
    
    def put_item(self,n,key,value):
        if n == None:
            return self.Node(key,value) #비어있는 노드라면 새로운 노드 생성
        if n.key > key:
            n.left = self.put_item(n.left,key,value) #노드가 없을 때까지 재귀적으로 반복
        if n.key < key:
            n.right = self.put_item(n.right,key,value)
        else:
            n.value = value #키가 이미 있을 때 value만 교체
        return n #밑에서부터 연결된 서브트리 반환
    
    def min(self):
        if self.root == None:
            return None
        return self.minimum(self.root)
    
    def minimum(self,n):
        if n.left == None:
            return n
        return self.minimum(n.left)
    
    def delete_min(self):
        if self.root == None:
            print('empty tree')
        self.root = self.del_min(self.root) #루트부터 탐색해 최솟값이 제외된 서브트리가 붙는다

    def del_min(self,n):
        if n.left == None: #n.left가 없으면 오른쪽 노드 리턴
            return n.right #오른쪽 노드가 None이어도 부모노드에는 None이 붙는다(=최소값이 삭제된다)
        n.left = self.del_min(n.left)
        return n
    
    def delete(self,k):
        if self.root == None:
            print('empty tree')
        self.root = self.del_node(self.root,k)

    def del_node(self,n,k):
        if n.key > k:
            n.left = self.del_node(n.left,k)
        elif n.key < k:
            n.right = self.del_node(n.right,k)
        else:
            if n.right == None: #오른쪽 노드가 없을 때
                return n.left
            if n.left == None: #왼쪽 노드가 없을 때
                return n.right
            
            target = n #자식이 둘다 있을 때

            n = self.minimum(target.right) # 타겟의 오른쪽 서브트리의 가장 최소값(유사 노드)

            n.right = self.del_min(target.right) #최소값이 삭제된 트리를 유사노드에 부착

            n.left = target.left

        return n #왼쪽 오른쪽 서브트리가 부착된 트리 반환            
```

## AVL 트리 (균형이진탐색트리)

- 균형이 깨질 때 균형을 맞춰주자! -> O(logN)을 보장함
- 균형이 맞지 않는다면 O(logN) -> O(N)으로 올라감
- A(h) : 높이가 h인 AVL트리를 구성하는 최소 노드 수
- 균형인수 : 왼쪽 서브트리 높이 - 오른쪽 서브트리 높이
    - tip: 각 노드를 부모노드라고 생각하고 트리의 좌우를 살피자

- 탐색연산 : 이진탐색트리와 동일
- 삭제와 삽입 : 불균형이 일어날때마다 균형을 맞춰줘야함
    - 균형을 맞추는 기준 노드 = 밑에서부터 올라가면 가장 처음 균형인수가 2인 조상노드
    - LL : 왼쪽+왼쪽 노드로 불균형 -> 오른쪽으로 회전
    - RR : 오른쪽+오른쪽 노드로 불균형 -> 왼쪽으로 회전
    - RL : 오른쪽+왼쪽 노드로 불균형 -> 오른쪽으로 회전 후 왼쪽 회전
    - LR : 왼쪽+오른쪽 노드로 불균형 -> 왼쪽으로 회전 후 오른쪽 회전

```python
class AVL:
    class Node:
            def __init__(self,key,value,height,left=None,right=None):
                self.key = key
                self.value=value
                self.height = height
                self.left = left
                self.right = right
                
    def __init__(self):
        self.root = None
    
    def height(self,n): #n의 높이
        if n == None:
            return 0 
        return n.height
    
    def rotate_right(self,n):
        x = n.left
        n.left = x.right
        x.right = n 
        n.height = max(self.height(n.left),self.height(n.right)) + 1
        x.height = max(self.height(x.left),self.height(x.right)) + 1
        return x
    
    def rotate_left(self,n):
        x = n.right
        n.right = x.left
        x.left = n
        n.height = max(self.height(n.left),self.height(n.right)) + 1
        x.height = max(self.height(x.left),self.height(x.right)) + 1
        return x
    
    def balance(self,n):
        if self.bf(n) > 1: # 균형인수가 2 이상일 때 ,왼쪽 서브트리가 더 높을 때
            if self.bf(n.left) < 0: #왼쪽서브트리의 오른쪽 서브트리가 더 높을 때 -> LR회전
                n.left = self.rotate_left(n.left)
            n = self.rotate_right(n) #아니면 RR회전만 수행
        if self.bf(n) < -1: #오른쪽 서브트리가 더 높을 때
            if self.bf(n.right) > 0: #오른쪽 서브트리의 왼쪽 서브트리가 더 높을 때 -> RL회전
                n = self.rotate_right(n.right)
            n = self.rotate_left(n) #아니면 LL회전만 수행 
        return n
    
    def bf(self,n):
        return self.height(n.left) - self.height(n.right)

    def put(self, key, value):
        self.root = self.put_item(self.root,key,value)

    def put_item(self, n, key, value):
        if n == None:
            return self.Node(key,value,1)
        if n.key > key:
            n.left = self.put_item(n.left,key,value)
        elif n.key < key:
            n.right = self.put_item(n.right,key,value)
        else:
            n.value = value # 키값이 같을 때 교체

        n.height = max(self.height(n.left),self.height(n.right)) + 1 #전체 트리의 높이 갱신
        return self.balance(n) # 균형을 잡은채로 리턴
    
    def delete(self,key):
        self.root = self.delete_node(self.root,key)

    def delete_node(self,n,key):
        if n == None:
            print('empty tree')
        
        if n.key > key:
            n.left = self.delete_node(n.left,key)
        elif n.key < key:
            n.right = self.delete_node(n.right,key)
        else:
            if n.left == None:
                return n.right
            if n.right == None:
                return n.left
            
            target = n
            n =  self.minimum(target.right)
            n.right = self.del_min(target.right)
            n.left = target.left

        n.height = max(self.height(n.left),self.height(n.right)) +1
        return self.balance(n)
    
    def min(self):
        if self.root == None:
            return None
        return self.minimum(self.root)
    
    def minimum(self,n):
        if n.left == None:
            return n
        return self.minimum(n.left)
    
    def delete_min(self):
        if self.root == None:
            print('empty tree')
        self.root = self.del_min(self.root) #루트부터 탐색해 최솟값이 제외된 서브트리가 붙는다

    def del_min(self,n):
        if n.left == None: #n.left가 없으면 오른쪽 노드 리턴
            return n.right #오른쪽 노드가 None이어도 부모노드에는 None이 붙는다(=최소값이 삭제된다)
        n.left = self.del_min(n.left)
        return n
```
