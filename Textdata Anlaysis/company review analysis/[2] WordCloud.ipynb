```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pickle
```

```python
font_path = 'C:\\Users\\user\\anaconda3\\envs\\text\\Lib\\site-packages\\matplotlib\\mpl-data\\fonts\\malgun.ttf'
```

```python
reviews = pd.read_csv('reviews_tokenized.csv',encoding='utf-8-sig')
```

```python
# 긍정과 부정으로 나누기 #
pos_review = reviews[reviews['label'] == 1]
neg_review = reviews[reviews['label'] == 0]
```

```python
# 긍정과 부정의 리뷰를 하나의 문자열로 합치기 #
pos_review = ' '.join(pos_review['text'].tolist())
neg_review = ' '.join(neg_review['text'].tolist())
```

```python
pos_cloud = WordCloud(font_path = font_path,background_color='white',
                  width=800, height=800).generate(pos_review)

pos_array  = pos_cloud.to_array() # 긍정 워드클라우드를 array로 변환
fig = plt.figure(figsize=(10, 10)) 
plt.imshow(pos_array) # 이미지 출력
plt.axis('off') # x y 축 숫자 제거
plt.show()
```

```python
neg_cloud = WordCloud(font_path = font_path,background_color='black',
                  width=800, height=800).generate(neg_review)

neg_array  = neg_cloud.to_array()
fig = plt.figure(figsize=(10, 10))
plt.imshow(neg_array)
plt.axis('off')
plt.show()
```
